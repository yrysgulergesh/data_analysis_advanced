```python
# 2

"""
В виде списка строк даны адреса. Достаньте из них данные и сохраните в отдельных переменных.

address = ['Бишкек, ул. Московская, д.15',

           'Бишкек, ул. Журнальная, д.45',
           'Токмак, ул. Центральная, д.31']

input:

address = ['Бишкек, ул. Московская, д.15', 'Бишкек, ул. Журнальная, д.45' ]

output:

city -> ['Бишкек', 'Бишкек']

street -> ['Московская', 'Журнальная']

number -> [15, 45]

"""
address = ['Бишкек, ул. Московская, д.15',
           'Бишкек, ул. Журнальная, д.45',
           'Токмак, ул. Центральная, д.31']

city = []
street = []
number = []


def get_addr(adr):
  l = adr.split()
  for i in range(len(l)):
    l[i] = l[i].strip(',.дул')
  l.pop(l.index(''))
  return l

print(get_addr('Бишкек, ул. Московская, д.15'))

for i in address:
  ad = get_addr(i)
  city.append(ad[0])
  street.append(ad[1])
  number.append(ad[2])

print(city)
print(street)
print(number)
```
