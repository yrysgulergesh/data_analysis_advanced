```python
n = int(input("Введите число n: "))

# Находим сумму чисел от 1 до n
sum_numbers = sum(range(1, n+1))

# Находим максимальное число от 1 до n
max_number = max(range(1, n+1))

# Находим минимальное число от 1 до n
min_number = min(range(1, n+1))

print("Сумма чисел от 1 до", n, ":", sum_numbers)
print("Максимальное число от 1 до", n, ":", max_number)
print("Минимальное число от 1 до", n, ":", min_number)
```

# Part 2

Найдем минимальное значение функции:

$$f(x) = x^2 - 2x + 5$$

```python
# Находим минимальное значение функции f(x) = x^2 - 2x + 5
# Вершина параболы находится в точке x = -b / (2a), где a = 1, b = -2
a = 1
b = -2

x_vertex = -b / (2*a)
y_min = a*x_vertex**2 - 2*x_vertex + 5

print("Минимальное значение функции f(x) = x^2 - 2x + 5 достигается при x =", x_vertex)
print("Минимальное значение функции:", y_min)
```

# Part 3

Найдем максимальное значение функции:

$$f(x) = -3x^2 + 12x - 4$$

```python
# Находим максимальное значение функции f(x) = -3x^2 + 12x - 4
# Вершина параболы находится в точке x = -b / (2a), где a = -3, b = 12
a = -3
b = 12

x_vertex = -b / (2*a)
y_max = a*x_vertex**2 + b*x_vertex - 4

print("Максимальное значение функции f(x) = -3x^2 + 12x - 4 достигается при x =", x_vertex)
print("Максимальное значение функции:", y_max)
```

# Part 6

Выполним расчет по формуле!

$$\sum_{i=1}^{20} \frac{x^2 -3}{x}$$

```python
round(sum([(i**2 - 3)/i for i in range(1, 21)]), 2)
```

```python
x = 1
result = 0

for i in range(1, 21):
    expression = (x ** 2 - 3) / x
    result += expression
    x += 1

print("Результат:", result)
```
