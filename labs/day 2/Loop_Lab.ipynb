Построение логических условий

Напишите программу, которая создает строку запрашивая ее ввод с клавиатуры.

Если ее длина от 3 до 7 символов, то программа выводит ее на печать.

В противном случае выводи первых 4 ее элемента

```python
user_input = input("Введите строку: ")

if 3 <= len(user_input) <= 7:
    print(user_input)
else:
    print(user_input[:4])
```

# Part 2

Напишите программу, которая создает строку запрашивая ее ввод с клавиатуры.

Если в строку входит латинский символ 'a', то программа меняет его на заглавный.

Если не входит, то возвращает сообщение:

"Символа "а" в строке нет"

```python
user_input = input("Введите строку: ")

if 'a' in user_input:
    modified_string = user_input.replace('a', 'A')
    print(modified_string)
else:
    print('Символа "а" в строке нет')
```

```python
def replicate_dna_sequence(dna_sequence):
    rna_sequence = ""
    for nucleotide in dna_sequence:
        if nucleotide == 'A':
            rna_sequence += 'U'
        elif nucleotide == 'G':
            rna_sequence += 'C'
        elif nucleotide == 'T':
            rna_sequence += 'A'
        elif nucleotide == 'C':
            rna_sequence += 'G'
    return rna_sequence

x = 'ATGTCCTATGTCCTAGCTATGTCCTAGCTGCGCAGCTATGTCCTAGCTGATGTCCTAGCTGCCGCTCGATATGTCCTAGCTGCGATGTAATGTCCTAGCTGCTGATGTCCTAGCTGCTCCTAGCTGCCCTAGCTGCTCGATG'
rna_sequence = replicate_dna_sequence(x)
print(rna_sequence)
```
