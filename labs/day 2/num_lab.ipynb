```python
# Найдем наименьшее общее кратное двух чисел
import math

def least_common_multiple(a, b):
    gcd = math.gcd(a, b)
    lcm = (a * b) // gcd
    return lcm

# Пример использования функции
num1 = 12
num2 = 18
result = least_common_multiple(num1, num2)
print("Наименьшее общее кратное:", result)
```
