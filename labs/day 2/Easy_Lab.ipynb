# Part 1

Напишите пожалуйста программу, которая получает на вход имя и возраст.

Программа возвращает два варианта сообщения:

'Здравствуйте, "Имя"!' для лиц сстарше 21 года, 

'Привет, "Имя"!' для остальных.

```python
name = input('Введите имя: ').capitalize()
age = int(input('Введите возраст: '))
if age >= 21:
  print('Здравствуйте, ' + name + '!')
else:
  print('')
```

# Part 2

На вход подается строка из первых трех букв английского алфавита. (Организуйте ее ввод с клавиатуры!)

Программа считает количество букв 'a'. если ее количество НЕЧЕТНО, то программа к введенной строке в конец приписывает (добавляет) еще одну букву 'a'. Если четно, то ничего со строкой не делает.

Затем программа возвращает введенную строку.


Если вы ввели:
'abcabac'

Ответ будет: 'abcabaca'

Если вы ввели:
'abcabaca'

Ответ будет: 'abcabaca'

```python
s = input()
s_correct = ''
for i in s:
  if (i=='a') or (i=='b') or (i=='c'):
    s_correct += i

num = s_correct.count('a')
if num%2 != 0:
  print(s_correct+'a')
else:
  print(s_correct)
```

# Part 3

Дана строка. Пробегитесь по ней циклом и после каждого символа 'A' добавте еще один такой же символ.

```python
x = 'ABCABBCCCBABCBBABCAB'
x_res = ''
for i in x:
  if i =='A':
    x_res = x_res + 'AA'
  else:
    x_res += i
print(x_res)
```

# Part 4

Дана строка чисел. Пробегитесь по ней циклом и составьте новую строку, в которую будут входить только четные цифры из исходной строки.

output:

'4682864286'

```python
num_str = '1465823758653428653'
res = ''
for i in num_str:
  if int(i)%2 == 0:
    res += i
res
```

# Part 5

Напишите программу, которая получает на вход 2 числа.

Данные числа являются ВКЛЮЧИТЕЛЬНЫМИ границами диапозона, из которого программа должна вернуть НЕЧЕТНЫЕ числа кратные трем!

Вам понадобиться цикл!

```python
a = int(input())
b = int(input())
res = ''
for i in range(a, b+1):
  if (i%2 !=0 ) and (i%3 == 0):
    res = res + str(i) + ' '
res
```
