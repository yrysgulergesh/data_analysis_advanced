## list generator

```python
%%time

# Version 1
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

```python
%%time

# Version 2 - Shorter and more elegant
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x  for x in fruits if "a" in x] 

print(newlist)
```

The Syntax

newlist = [*item* **for** *item* **in** *iterable*]

```python
x = [i**2 for i in range(11)]
y = [str(i) for i in range(21)]

s = 'jhsdkjhfljkashdlkjhlkjdfhglkjsdfh'
l = [i for i in s]

print(l)
```
