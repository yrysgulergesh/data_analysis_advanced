В них возникает необходимость, когда в зависимости от одного или нескольких условий есть несколько возможных сценариев действий. Их может быть сколь угодно много.



*   if - ключевое слово, которое задает главное условие (основной сценарий)
*   else - задает альтернативный сценарий, если не выполняется главное условие

*   elif - (else if) - задает дополнительные условия, если возможных сценариев больше чем 2





```python
y = 'AHJDHSGA'
if 'f' in y:
  print('Yes')
else:
  print('No')
```

```python
# Как видим, иногда нет альтернативного действия

if y.count('J') >= 2:
  y = y + '_COOL'
  print(y)
```

Циклы. Это одна из фундаментальных функций программной среды в контексте написания алгоритмов. Идея итерации - одна из ключевых!

```python
for i in range(4, 1, -1):
  print(i)
```

```python
# выведем все числа, которые кратны трем от 1 до 50
for i in range(-5, 51):
  if i%3 == 0:
    print(i, end='    ')
```

```python
# Эту же задачу можно (в теории) решить с помощью цикла while:
# но данный цикл на практике почти никогда не применяется, ибо он стремный и медленный. 
# Более того, даже использования цикла for на практике стоит избегать, хотя иногда он нужен, но никак не while
n = 0
while n < 50:
  n = n+3
  print(n)
```

## Решим квадратное уравнение методом подбора
При условии, что корни - целые числа

$$x^2 - 8x + 12 = 0$$

```python
for i in range(-10000, 10000):
  if (i**2 - i*8 + 12)==0:
    print(i)
```

```python
# кастомизируем программу
x_1 = str()
x_2 = str()
for i in range(-100, 100):
  if (i**2 - i*8 + 12)==0:
    if len(x_1) == 0:
      x_1 = x_1 + str(i)
    else:
      x_2 = x_2 + str(i)
  if (len(x_1) != 0) and (len(x_2) != 0):
    break

print('Корни уравнения: ' + x_1 + ' и ' + x_2 + '.')
```

### Применим цикл в работе со строками

```python
a = 'abcdABCD1234'
for i in a:
  print(i, end=' ')
```

```python
# найдем максимальную длину фрагмента из следующих друг за другом повторяющихся элементов
x = 'ABCCBBAAABCACAAAAAAAAABBCABACAAAABBACCCCCAABBBBBBBBCCCACBBBBCB'
count = str()
n = 1
for i in range(1, len(x)):
  if x[i] == x[i-1]:
    n += 1
  else:
    count += str(n) + ' '
    n = 1
count


```

```python
res = 0
for i in range(0, len(count), 2):
  if res < int(count[i]):
    res = int(count[i])
res
```
