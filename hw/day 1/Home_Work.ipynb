# Задача 2

Напишите программу, которая получает строку при вводе с клавиатуры и возвращает количество введенных символов. (Используйте input()).

Задача 4

Напишите программу, которая получает с клавиатуры 2 строки и сравнивает их длины. Программа должна вернуть строку:
"Строки отличаются на 'N' символов".
Вам может пригодиться функция abs() которая всегда возвращает положительное число. Это нужно, чтобы вы не получили отрицательную разность!

```python
string1 = input("Введите первую строку: ")
string2 = input("Введите вторую строку: ")

length_difference = abs(len(string1) - len(string2))
result_string = f"Строки отличаются на '{length_difference}' символов."
print(result_string)
```

# Задача 5

Давайте потренеруем слайсинг)

В данной строке выведите следущие элементы:


---



*   3-й символ

*   Первые 2

*   Последние 3
*   3-й и 2-й с конца


*   Всю строку через 1 символ


*   2-й и последний символ


*   Всю строку в обратном порядке
*   С 2-го по предпоследний

*   С 3-го по 6-й включительно
*   3 последних начиная с конца***







```python
s = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890'

# 3-й символ
print(s[2])

# Первые 2
print(s[:2])

# Последние 3
print(s[-3:])

# 3-й и 2-й с конца
print(s[-3:-5:-1])

# Вся строка через 1 символ
print(s[::2])

# 2-й и последний символ
print(s[1] + s[-1])

# Вся строка в обратном порядке
print(s[::-1])

# С 2-го по предпоследний
print(s[1:-1])

# С 3-го по 6-й включительно
print(s[2:7])

# 3 последних начиная с конца
print(s[-3:])
```

```python
s = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890'

# 3rd character
print(s[2])

# First 2 characters
print(s[:2])

# Last 3 characters
print(s[-3:])

# 3rd and 2nd characters from the end
print(s[-3:-5:-1])

# Every other character
print(s[::2])

# 2nd and last character
print(s[1] + s[-1])

# Reverse the string
print(s[::-1])

# From the 2nd to the second-to-last character
print(s[1:-1])

# From the 3rd to the 6th character (inclusive)
print(s[2:7])

# Last 3 characters from the end
print(s[-3:])
```
