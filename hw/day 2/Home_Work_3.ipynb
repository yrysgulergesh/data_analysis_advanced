## Задача 9

Дана строка. Попробуйте определить, сколько раз встречается самый частый элемент в строке.

input: 'ABBCCCCA'

output: 4

```python
def count_most_frequent(string):
    char_count = {}
    max_count = 0
    
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
            
        if char_count[char] > max_count:
            max_count = char_count[char]
    
    return max_count

# Пример использования:
string = 'ABBCCCCA'
result = count_most_frequent(string)
print(result)  # Вывод: 4

# Другой пример:
s = 'ABCBBCDBBADBBBBCDBBBBCBDABBBBCDBBBABDBCBBBBDCABADCBCBADCBCBBBCDB'
result = count_most_frequent(s)
print(result)  # Вывод: 12


```

## Задача 10 *

Продолжите предыдущую задачу и попытайтесь найти самый часто встречающийся элемент.

input: 'ABBCCCCA'

output: 'C'

```python
def find_most_frequent(string):
    char_count = {}
    max_count = 0
    most_frequent_char = ''
    
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
            
        if char_count[char] > max_count:
            max_count = char_count[char]
            most_frequent_char = char
    
    return most_frequent_char

# Пример использования:
string = 'ABBCCCCA'
result = find_most_frequent(string)
print(result)  # Вывод: 'C'

# Другой пример:
s = 'ABCBBCDBBADBBBBCDBBBBCBDABBBBCDBBBABDBCBBBBDCABADCBCBADCBCBBBCDB'
result = find_most_frequent(s)
print(result)  # Вывод: 'B'
```

## Задача 11**

Даны 2 строки - одна с символами всех клавиш компьютора в регистре Кириллица, другая с символами Латиницы.

По Этим двум строками соберите новую строку, где будут идти друг за другом элементы одной клавиши для всех элементов

'qйwцeуrкtеyнuгiшoщpз.....'

Используйте цикл, элементами которого будут индексы строк!

```python
def merge_strings(cyrillic_string, latin_string):
    merged_string = ""
    for i in range(max(len(cyrillic_string), len(latin_string))):
        if i < len(cyrillic_string):
            merged_string += cyrillic_string[i]
        if i < len(latin_string):
            merged_string += latin_string[i]
    return merged_string

# Пример использования:
cyrillic_string = "йцукенгшщзхъфывапролджэячсмитьбю"
latin_string = "qwertyuiopasdfghjklzxcvbnm"
result = merge_strings(cyrillic_string, latin_string)
print(result)  # Выводит 'qйwцeуrкtеyнuгiшoщpз....'
```
