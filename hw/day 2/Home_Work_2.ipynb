## Задача 3.

Создайте переменную, которая на вход получает четырехзначное число через input().

Напишите программу, которая строит по заданному числу новое двухзначное по следущему принципу:

Первая цифра - это разность двух первых цифр введенного числа, а второе соответственно есть разность цифр последних двух цифр введенного.

Чтобы разность получалась всегда положительным числом используйте функцию abs().

пример: 

abs(-7)

  7

Выведите полученное число

```python
# Ваш код:
number = int(input("Введите четырехзначное число: "))
first_digit = abs((number // 1000) - ((number // 100) % 10))
second_digit = abs((number % 100) // 10 - (number % 10))
result = first_digit * 10 + second_digit
print(result)
```

```python
num = int(input("Введите четырехзначное число: "))
first_digit = abs((num // 1000) - ((num // 100) % 10))
second_digit = abs((num % 100) // 10 - (num % 10))
result = first_digit * 10 + second_digit
print("Новое двухзначное число:", result)


```

Создайте переменную, которое на вход получает 3х, 4х, 5ти...-значное число.


Если последние 2 цифры вместе являются четным числом, замените их на 2.

В противном случае оставьте его как есть.

```python
# Ваш код:
x = input()
temp = int(x[-2:])
if temp%2 == 0:
  x = x[:-2] + '2'
  print(x)
else:
  print(x)


```

```python
number = input("Введите число: ")
last_two_digits = int(number[-2:])

if last_two_digits % 2 == 0:
    modified_number = number[:-2] + "2"
else:
    modified_number = number
    print("Модифицированное число:", modified_number)
```

## Задача 6.

На вход подается число. (Вы уже знаете, каким образом)

Если число меньше или равно 5 или больше 7ми, то возмите его квадратный корень и верните результат.

В противном случае поделите его на 2 и выведите результат

```python
# Ваш код:
number = float(input("Введите число: "))

if number <= 5 or number > 7:
    result = number ** 0.5
else:
    result = number / 2

print("Результат:", result)
```
