Обратите внимание на разные скобки, используемые для обозначения интервалов. В задании используются полуоткрытые и открытые интервалы. Подробнее про это вы можете прочитать, например, на википедии (полуинтервал, промежуток).


```python
def check_interval(num):
    if (-15 < num <= 12) or (14 < num < 17) or (num >= 19):
        return True
    else:
        return False

# Пример использования программы
number = int(input("Введите целое число: "))
result = check_interval(number)
print(result)
```

# Задача 2

Дана строка с символами.

Напишите программу, которая на вход получает символ (в любом регистре) и считает, сколько раз данный символ встречается в строке. Если такого символа в строке нет, то сообщает об этом. 

Пример:

input: 'A'

output: 'Символ "A" встречается в строке 6 раз.'

input: 'Z'

output: 'Символ "A" в строке не встречается.'

```python

my_str = 'ABCCBBABDBCBBFBABBABDFFAAD'
char = input('Введите символ: ')

count = my_str.count(char)
if count > 0:
    print(f'Символ "{char}" встречается в строке {count} раз.')
else:
    print(f'Символ "{char}" в строке не встречается.')
```

# Задача 3

Напишите программу, которая получает на ввод строку латинских символов, делает эту строку исключительно из строчных (маленьких) букв и возвращает строку только уникальных значений (без повторов)

input: 'Aabbbcdccbacd'

output: 'abdc'

```python
string = input('Введите строку: ')
string = string.lower()
unique_chars = ''.join(set(string))

print(unique_chars)
```
