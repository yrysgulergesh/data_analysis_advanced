# Task 4

Создайте 2 произвольных списка (строковых). Напишите функцию, которая возвращает все элементы, которые встречаются в обоих списках

````python
def find_common_elements(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    common_elements = set1.intersection(set2)
    return list(common_elements)

# Пример использования функции
a = [i for i in 'jhjhjkjghf']
b = [i for i in 'hjfurlkk']
result = find_common_elements(a, b)
print("Элементы, встречающиеся в обоих списках:", result)
````

# Task 6

Даны 2 предложения.

Напишите функцию, которая исправит в них ошибки.

````python
x_1 = 'Клиент иван гРОМОВ внес, 500 сом в качестве первоначального платежа,'
x_2 = 'Меня зовут сергей петров и я алкоголик@'
````

````python
def fix_errors(sentence):
    # Удаление пробелов перед запятыми и точками
    sentence = sentence.replace(" ,", ",")
    sentence = sentence.replace(" .", ".")
    
    # Замена первой буквы на заглавную
    sentence = sentence.capitalize()
    
    # Исправление имени и фамилии: первая буква заглавная, остальные строчные
    words = sentence.split()
    for i in range(len(words)):
        if words[i].isalpha():
            words[i] = words[i].capitalize()
    
    # Склеивание слов обратно в предложение
    fixed_sentence = " ".join(words)
    
    return fixed_sentence

# Пример использования функции
x_1 = 'Клиент иван гРОМОВ внес, 500 сом в качестве первоначального платежа,'
x_2 = 'Меня зовут сергей петров и я алкоголик@'

fixed_x_1 = fix_errors(x_1)
fixed_x_2 = fix_errors(x_2)

print("Исправленное предложение 1:", fixed_x_1)
print("Исправленное предложение 2:", fixed_x_2)
````

# Task 7

Напишите функцию, которая вычисляет самму, разность и произведение двух чисел и возвращает список результатов.

````python
def calculate_operations(num1, num2):
    sum_result = num1 + num2
    difference_result = num1 - num2
    product_result = num1 * num2
    results = [sum_result, difference_result, product_result]
    return results

# Пример использования функции
num1 = 10
num2 = 5
result = calculate_operations(num1, num2)
print("Результаты операций:", result)
````

Напишем простенькую функцию, которая получает на вход двумерный список и красиво его выводит на печать

````python
array = [[1, 22, 3], [4, 5, 66], [77, 8, 9]]
array[1]
````

````python
def pr(x):
  for i in x:
    for j in i:
      if len(str(j))==1:
        print(j, end='  ')
      else:
        print(j, end=' ')
    print()

````

````python
pr(array)
````

Теперь мы можем в дальнейшем использовать эту функцию для отображения двумерных списков.

````python
def calculate_percentage_difference(num1, num2):
    difference = num2 - num1
    percentage_difference = (difference / num1) * 100
    return percentage_difference

def display_2d_list(list_2d):
    for row in list_2d:
        for num in row:
            print(num, end='\t')
        print()

# Пример использования функции
list_2d = [[50, 75, 80], [30, 40, 20], [100, 90, 95]]

# Применяем функцию к каждой паре чисел в двумерном списке
for row in list_2d:
    for i in range(len(row)-1):
        num1 = row[i]
        num2 = row[i+1]
        result = calculate_percentage_difference(num1, num2)
        print("Разница в процентах между", num1, "и", num2, ":", result)

# Отображаем двумерный список
display_2d_list(list_2d)
````

```
[[price, squere], [price, squere]...]
```

$$prise = prise-0.5(prise-105sq)$$

Функция должна вернуть одномерный массив со стоимостями.



````python
x = [[30000, 105], [25000, 100], [3500, 200], [60000, 200], [300000, 250], [12, 208]]
````

````python
def adjust_prices(list_2d):
    adjusted_prices = []
    for row in list_2d:
        price = row[0]
        square = row[1]
        adjusted_price = price - 0.5 * (price - 105 * square)
        adjusted_prices.append(adjusted_price)
    return adjusted_prices

# Пример использования функции
x = [[30000, 105], [25000, 100], [3500, 200], [60000, 200], [300000, 250], [12, 208]]
adjusted_prices = adjust_prices(x)
print("Скорректированные стоимости:", adjusted_prices)
````

Напишите функцию, которая получает на вход размер квадратной матрицы (количество цифр в строке) и возвращает единичную матрицу.

````python
def ones(size):
    matrix = []
    for i in range(size):
        row = []
        for j in range(size):
            if i == j:
                row.append(1)
            else:
                row.append(0)
        matrix.append(row)
    return matrix

# Пример использования функции
size = 8
result = ones(size)
for row in result:
    print(row)
````
